##Broadcasting

Because of index alignment, you will want to make sure that the indexes:
*   Are unique (no duplicates)
*   Are common to both series

```python
s1 = pd.Series([10, 20, 30], index=[1,2,2])
s2 = pd.Series([35, 44, 53], index=[2,2,4], name='s2')
```

```python
s1
```

```python
s2
```

```python
s1 + s2
```
